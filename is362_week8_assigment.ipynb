### 1. Load the data in the (yellow highlighted) Data Folder in the file auto-mpg.data into a pandas DataFrame. 

```python
auto_mpg = pd.read_fwf("https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data",header=None)
auto_mpg
```

### 3. Use the attribute information provided in auto-mpg.names to name your pandas DataFrame columns. 

```python
auto_mpg = auto_mpg.reset_index()
auto_mpg.rename(columns={0:'mpg',1:"cylinders",2:"displacement",3:"horsepower",4:"weight",5:"acceleration",6:"model year",7:"origin",8:"car name"}, inplace=True)
auto_mpg
```

### 4. Some of the horsepower data is marked as missing with a â€˜?â€™.  You need to replace this â€˜?â€™ with a pandas-appropriate missing value, then convert the column to numeric.

```python
auto_mpg=auto_mpg.replace('?',np.nan)
auto_mpg
```

```python
auto_mpg= auto_mpg.fillna(0)
auto_mpg
```

### 5. The origin column values are 1, 2, and 3.  Convert these values respectively to â€˜USAâ€™, â€˜Asiaâ€™, and â€˜Europeâ€™ 

```python
auto_mpg['origin'].replace([1],'USA', inplace=True)
auto_mpg['origin'].replace([2],'Asia', inplace=True)
auto_mpg['origin'].replace([3],'Europe', inplace=True)
auto_mpg
```
